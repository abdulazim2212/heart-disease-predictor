# Data Preprocessing

```python
df.isna().sum()
```

```python
df['trestbps'] = df['trestbps'].fillna(df['trestbps'].mean())
df['chol'] = df['chol'].fillna(df['chol'].mean())
df['fbs'] = df['fbs'].fillna(df['fbs'].mode()[0])
df['restecg'] = df['restecg'].fillna(df['restecg'].mode()[0])
df['thalch'] = df['thalch'].fillna(df['thalch'].mean())
df['exang'] = df['exang'].fillna(df['exang'].mode()[0])
df['oldpeak'] = df['oldpeak'].fillna(df['oldpeak'].mean())
df['slope'] = df['slope'].fillna(df['slope'].mode()[0])
df['ca'] = df['ca'].fillna(df['ca'].mean())
df['thal'] = df['thal'].fillna(df['thal'].mode()[0])
df['num'] = df['num'].fillna(df['num'].mode()[0])
# df = df.dropna()
```

```python
df.isna().sum()
```

```python
# Log Transformation of Serum Cholesterol Levels
df['chol'] = np.log(df['chol'] + 1)
```

```python
# Distribution of Serum Cholesterol Levels (chol) after log transformation
sns.distplot(df['chol'])
plt.title('Distribution of Serum Cholesterol Levels (chol)')
plt.show()
```

```python
# Log Transformation of Resting Blood Pressure (trestbps)
df['trestbps'] = np.log(df['trestbps'])
```

```python
# Boxplot of Resting Blood Pressure (trestbps) after log transformation
sns.boxplot(df['trestbps'])
plt.title('Boxplot of Resting Blood Pressure (trestbps)')
plt.show()
```

```python
# df1 = df.drop(['id','thalch','chol'],axis=1) 78%precision on dtc
df1 = df.drop(['id','dataset','thalch','chol'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
for col in df1.columns.tolist():
    le = LabelEncoder()
    df1[col]= le.fit_transform(df1[col])
```

```python
df1.shape
```

```python
x = df1.drop('num',axis=1)
y= df1['num']
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import classification_report
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=0)  
def classify(model):
    model.fit(x_train,y_train)
    pr = model.predict(x_test)
    cr = classification_report(y_test,pr)
    print(cr)
    print("Accuracy : ",model.score(x_test,y_test))
    score = cross_val_score(model,x,y,cv=5)
    print("CV score",score.mean())
```

```python
from sklearn import svm
model = svm.SVC(kernel='linear')
classify(model)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
classify(model)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc_model = DecisionTreeClassifier()
classify(model)
```

```python
import tensorflow as tf
dtc_model.fit(x_train,y_train)
pred = dtc_model.predict(x_test)
cm = tf.math.confusion_matrix(y_test,pred)
sns.heatmap(cm,annot=True,cmap='coolwarm')
```

```python
from sklearn.ensemble import BaggingClassifier
model = BaggingClassifier(
base_estimator=DecisionTreeClassifier(),
    n_estimators=100,
    max_samples=0.80,
    oob_score=True,
    random_state=0
)
classify(model)
```

```python
model.fit(x_train,y_train)
y_prediction= model.predict(x_test)
cm = tf.math.confusion_matrix(y_test,y_prediction)
sns.heatmap(cm,annot=True,cmap='coolwarm')
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
classify(model)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
classify(model)
```

```python
from catboost import CatBoostClassifier
model = CatBoostClassifier(verbose=0)
classify(model)
```

```python
from xgboost import XGBClassifier
model =XGBClassifier()
classify(model)
```
